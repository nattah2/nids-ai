```python
# Test to calculate time since last packet sent from machine
time_since = []
last_seen = {}
for packet in df.itertuples():

    if (packet[2] in last_seen):
        time_since.append(packet[1] - last_seen[packet[2]])
    else:
        time_since.append(0)
    last_seen[packet[2]] = packet[1]
up_df = df.copy()
up_df['Time Since Last'] = time_since
```

```python
df['TCP Flags'].value_counts()
cat_df = df.copy()

#       URG ACK PSH RST SYN FIN
flags = [0,  0,  0,  0,  0,  0]
indexes = [[], [], [], [], [] ,[]]
is_used = [[], [], [], [], [], []]
for index, flag_int in enumerate(df['TCP Flags']):
    flag_bin = bin(flag_int)

    get_flags = [0, 0, 0, 0, 0, 0]
    
    try:
        if (int(flag_bin[-1])):
            flags[5] += 1
            indexes[5].append(index)
            get_flags[5] = 1
        if (int(flag_bin[-2])):
            flags[4] += 1
            indexes[4].append(index)
            get_flags[4] = 1
        if (int(flag_bin[-3])):
            flags[3] += 1
            indexes[3].append(index)
            get_flags[3] = 1
        if (int(flag_bin[-4])):
            flags[2] += 1
            indexes[2].append(index)
            get_flags[2] = 1
        if (int(flag_bin[-5])):
            flags[1] += 1
            indexes[1].append(index)
            get_flags[1] = 1
        if (int(flag_bin[-6])):
            flags[0] += 1
            indexes[0].append(index)
            get_flags[0] = 1
    except:
        1 == 1
    finally:
        is_used[0].append(get_flags[0])
        is_used[1].append(get_flags[1])
        is_used[2].append(get_flags[2])
        is_used[3].append(get_flags[3])
        is_used[4].append(get_flags[4])
        is_used[5].append(get_flags[5])
```

```python
cat_df['URG'] = is_used[0]
cat_df['ACK'] = is_used[1]
cat_df['PSH'] = is_used[2]
cat_df['RST'] = is_used[3]
cat_df['SYN'] = is_used[4]
cat_df['FIN'] = is_used[5]
cat_df
```

```python
sns.scatterplot(data=cat_df, x='TCP Flags', y='Destination Port')
plt.xlim(1, 20) # 0 means no flags -- not relevant to analysis
plt.show()
```

```python
plt.bar(['URG', 'ACK', 'PSH', 'RST', 'SYN', 'FIN'], flags)
plt.title('TCP Flags')
plt.show()
```
